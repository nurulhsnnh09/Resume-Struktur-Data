STACK

Stacks adalah satu struktur data dimana penambahan dan penghapusan data, hanya dapat
dilakukan pada satu ujung yang sama, atau yang biasa dikenal dengan istilah top.
Semakin data jauh berada dari posisi top, maka data tersebut diindikasikan berada di stack lebih
lama dibandingkan dengan data yang berada dekat pada data di posisi top.
Jika terdapat data baru yang ditambahkan di stack, maka data ini pulalah yang akan dihapus
ketika terdapat proses penghapusan data. Konsep ini dikenal dengan nama LIFO-Last In First
Out.
Stack juga adalah struktur data linier yang mengikuti prinsip LIFO (Last In, First Out), di mana elemen terakhir yang dimasukkan adalah yang pertama kali dikeluarkan. Konsep ini mirip tumpukan piring: item baru ditambahkan di atas (push) dan item teratas diambil terlebih dahulu (pop). Stack sering digunakan untuk manajemen fungsi, undo/redo, dan evaluasi ekspresi matematika.

Implementasi-Queues

Contoh implementasi queues ini adalah permainan anak-anak, yaitu permainan ular naga. Pada

Berikut code simulasi dari permainan tersebut, dengan sedikit modifikasi, yaitu fungsi penjaga dan
lagu 'ular naga' digantikan dengan bilangan tertentu, dan sekolompok anak disusun dalam suatu
antrian. Permainan akan terus berjalan, sampai dengan hitungan bilangan tertentu tersebut. Jika
pada hitungan tertentu, terdapat anak yang berada di ujung antrian (front), maka anak tersebut
harus keluar barisan antrian. Permainan dimenangkan oleh seorang anak yang tersisa di dalam

Code Program :

```python
# fungsi queue
def createQueue():
    return []
def enqueue(q, data):
    q.append(data)
def dequeue(q):
    return q.pop(0)
def size(q):
    return len(q)

# fungsi ular naga
def ularNaga(nama, hitungan):
    gameQueue = createQueue()
    
    # memasukkan semua nama ke queue
    for namaAnak in nama:
        enqueue(gameQueue, namaAnak)
    
    print('Peserta Permainan =', gameQueue)
    
    # proses permainan
    while size(gameQueue) > 1:
        for i in range(hitungan):
            enqueue(gameQueue, dequeue(gameQueue))
            print('hitungan ke-', i+1, '=', gameQueue)
        
        # keluarkan 1 orang
        keluar = dequeue(gameQueue)
        print('Yang keluar =', keluar)
        print('Peserta Permainan =', gameQueue)
    
    return dequeue(gameQueue)
    
# pemanggilan
pemenang = ularNaga(['andi','rita','sari','anton','rafa','diana','zaki'], 2)
print("Pemenang:", pemenang)
```

implementasi sederhana antrean pelanggan menggunakan struktur data queue pada bahasa pemrograman Python.

```python
class Queue:
   def __init__(self):
       self.items = []
 
   def is_empty(self):
       return len(self.items) == 0
 
   def enqueue(self, item):
       self.items.append(item)
 
   def dequeue(self):
       if not self.is_empty():
           return self.items.pop(0)
       else:
           return "Queue is empty"
 
   def size(self):
       return len(self.items)
 
# Contoh penggunaan
antrian = Queue()
 
# Menambahkan elemen ke dalam antrian
antrian.enqueue("Pelanggan 1")
antrian.enqueue("Pelanggan 2")
antrian.enqueue("Pelanggan 3")
 
# Melihat ukuran antrian
print("Ukuran antrian:", antrian.size())
 
# Mengambil elemen dari antrian
print("Panggilan ke kasir:", antrian.dequeue())
 
# Melihat ukuran antrian setelah panggilan pertama
print("Ukuran antrian setelah panggilan pertama:", antrian.size())
```

Infix

Bentuk umum:

Prefix

Prefix adalah penulisan operasi di mana operator berada di depan operand.

Infix -> Postfix

```python
def pr(op):
    if op in "+-":
        return 1
    if op in "*/":
        return 2
    return 0

def ke_postfix(exp):
    s = []
    hasil = ""

    for i in exp:
        if i.isalnum():
            hasil += i
        elif i == '(':
            s.append(i)
        elif i == ')':
            while s[-1] != '(':
                hasil += s.pop()
            s.pop()
        else:
            while s and pr(i) <= pr(s[-1]):
                hasil += s.pop()
            s.append(i)

    while s:
        hasil += s.pop()

    return hasil

data = input("Infix: ")
print("Postfix:", ke_postfix(data))
```

Infix -> Prefix

```python
def pr(op):
    if op in "+-":
        return 1
    if op in "*/":
        return 2
    return 0

def ke_postfix(exp):
    s = []
    hasil = ""

    for i in exp:
        if i.isalnum():
            hasil += i
        elif i == '(':
            s.append(i)
        elif i == ')':
            while s[-1] != '(':
                hasil += s.pop()
            s.pop()
        else:
            while s and pr(i) <= pr(s[-1]):
                hasil += s.pop()
            s.append(i)

    while s:
        hasil += s.pop()

    return hasil


def ke_prefix(exp):
    exp = exp[::-1]

    baru = ""
    for i in exp:
        if i == '(':
            baru += ')'
        elif i == ')':
            baru += '('
        else:
            baru += i

    hasil = ke_postfix(baru)
    return hasil[::-1]


data = input("Infix: ")
print("Prefix:", ke_prefix(data))
```

Implementasi Algoritma Bubble Sort

```python
def bubble_sort(data):
    n = len(data)

    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j+1]:
                data[j], data[j+1] = data[j+1], data[j]

    return data


data = list(map(int, input("Masukkan angka: ").split()))

print("Hasil sort:", bubble_sort(data))
```

```python
def bubble_sort(data):
    n = len(data)

    for i in range(n):
        tukar = False  # penanda

        for j in range(0, n - i - 1):
            if data[j] > data[j+1]:
                data[j], data[j+1] = data[j+1], data[j]
                tukar = True  # ada pertukaran

        if tukar == False:
            break  # berhenti kalau sudah rapi

    return data


data = [1, 2, 3, 4, 5]
print("Hasil:", bubble_sort(data))
```

```python
def bubble_dua_arah(data):
    kiri = 0
    kanan = len(data) - 1

    while kiri < kanan:
        # kiri ke kanan
        for i in range(kiri, kanan):
            if data[i] > data[i+1]:
                data[i], data[i+1] = data[i+1], data[i]

        kanan -= 1

        # kanan ke kiri
        for i in range(kanan, kiri, -1):
            if data[i] < data[i-1]:
                data[i], data[i-1] = data[i-1], data[i]

        kiri += 1

    return data


data = [5, 2, 4, 1, 3]
print("Hasil:", bubble_dua_arah(data))
```

Algoritma Selection Sort

Algoritma selection sort memiliki waktu komputasi yang lebih cepat dibandingkan dengan
algoritma bubble sort. 

Hal ini dikarenakan terjadi reduksi waktu pada saat proses pertukaran
data. Jika algoritma bubble sort, pada setiap iterasi,setiap dua data yang berdekatan akan
dibandingkan, dan akan dilakukan pertukaran data jika syarat memenuhi.

Maka pada algoritma selection sort tidak akan melakukan dua proses ini (perbandingan dan
pertukaran) pada satu waktu, akan tetapi, pada algoritma selection sort, akan dicari data yang
memenuhi syarat terlebih dahulu (perbandingan data), kemudian data yang terpilih ini ini akan
diletakkan pada indeks data yang tepat. Oleh karena itu algoritma ini dinamakan selection
(pemilihan data).

Berikut tahapan algoritma selection sort (ascending order):

1. Algoritma ini dimulai dari indeks awal sampai dengan indeks akhir data
   
2. Cari data dengan nilai paling minimal (dari indeks awal sampai dengan indeks akhir) melalui
proses perbandingan


3. Letakkan data minimal ini di indeks awal

   
4. Ulangi lagi proses pencarian data paling minimal (dari indeks awal+1 sampai dengan indeks
akhir, karena indeks awal sudah terisi data yang tepat).


5. Letakkan data ini pada indeks awal+1

   
6. Ulangi lagi proses pencarian data paling minimal (dari indeks awal+2 sampai dengan akhir)
dan letakkan data ini pada indeks awal+2, dst.

```python
import random
import time

# buat data random 500 angka
data = [random.randint(1, 1000) for i in range(500)]

# bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                # tukar
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# selection sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # tukar
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# copy data biar adil
data1 = data.copy()
data2 = data.copy()

# waktu bubble sort
start = time.time()
bubble_sort(data1)
end = time.time()
print("Waktu Bubble Sort:", end - start)

# waktu selection sort
start = time.time()
selection_sort(data2)
end = time.time()
print("Waktu Selection Sort:", end - start)
```

2. Modifikasi Selection Sort (cari min & max sekaligus)

```python
def selection_sort_modif(arr):
    kiri = 0
    kanan = len(arr) - 1

    while kiri < kanan:
        min_idx = kiri
        max_idx = kiri

        for i in range(kiri, kanan + 1):
            if arr[i] < arr[min_idx]:
                min_idx = i
            if arr[i] > arr[max_idx]:
                max_idx = i

        # tukar nilai minimum ke kiri
        arr[kiri], arr[min_idx] = arr[min_idx], arr[kiri]

        # kalau max ketukar posisi, update index
        if max_idx == kiri:
            max_idx = min_idx

        # tukar nilai maksimum ke kanan
        arr[kanan], arr[max_idx] = arr[max_idx], arr[kanan]

        kiri += 1
        kanan -= 1

    return arr
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1

        arr[j+1] = key
    return arr

data = [5, 2, 8, 1]
print(insertion_sort(data))
```

```python
def cek_palindrome(kata):
    # balik kata
    balik = kata[::-1]

    # cek sama atau tidak
    if kata == balik:
        return "Palindrome"
    else:
        return "Bukan palindrome"

# contoh
teks = input("Masukkan kata: ")
print(cek_palindrome(teks))
```

Program Palindrome Pakai Stack

```python
def cek_palindrome(kata):
    stack = []

    # hapus spasi dan samakan huruf
    kata = kata.replace(" ", "").lower()

    # masukin ke stack
    for huruf in kata:
        stack.append(huruf)

    # ambil dari stack (dibalik)
    balik = ""
    while stack:
        balik += stack.pop()

    # bandingkan
    if kata == balik:
        return "Palindrome"
    else:
        return "Bukan palindrome"

# contoh
teks = input("Masukkan kata: ")
print(cek_palindrome(teks))
```

Penjelasan :

stack itu sifatnya LIFO (terakhir masuk, keluar duluan)

pas di-pop, otomatis kebalik

hasilnya dibandingin sama kata asli

Iteratif 

Iteratif merupakan penyelesaian permasalahan dengan perulangan, menggunakan syntax for atau while (pada Python).

Misalkan permasalahan perhitungan faktorial suatu bilangan yaitu :

5! = 5 x 4 x 3 x 2 x 1, atau n! = n x (n-1) x (n-3)x ... x1

maka perhitungan faktorial suatu bilangan dapat dilakukan dengan cara iterasi, yaitu menghitung operasi perkalian mulai dari bilangan tersebut sampai dengan satu.

code untuk perhitungan faktorial suatu bilangan dengan menggunakan cara iteratif.
Pada code ini, terdapat iterasi while, yang menghitung perkalian, mulai dari , sampai dengan 2.

```python
def factorialIteratif(bilangan):

    if bilangan <=1:
       return(1)
    else:
        temp=bilangan
        hasil=1
        while temp>1:
            hasil=hasil*temp
            temp=temp-1
        return(hasil)

print(factorialIteratif(4))
```

Suatu permasalahan akan lebih mudah dipecahkan jika permasalahan tersebut berukuran kecil.

```python
def triangularNumbers1(n):
    total = 0
    temp = n
    while temp>=1:
        total+=temp
        temp-=1
    return(total)
print(triangularNumbers1(2))
```

pindahkah piringan sebanyak n-1 dari tiang awal ke tiang bantuan, melalui tiang tujuan

dibawah ini code untuk penyelesaian permasalahan towers of hanoi dengan menggunakan teknik rekursif


```python
def towers(n,awal,bantuan,tujuan):
    if n==1:
        print("Piringan - 1 dari-", awal, "ke-",tujuan)
    else:
        towers(n-1,awal,tujuan,bantuan)
        print("Piringan -",n, "dari-",awal,"ke-",tujuan)
        towers(n-1,bantuan,awal,tujuan)
towers(4,'A','B','C')
```

Algoritma yang pertama adalah sequential search.

Sequential Search merupakan teknik pencarian data dengan cara membandingkan data yang
dicari dengan seluruh data yang terdapat pada kumpulan data secara satu persatu, mulai dari data
pertama sampai dengan data terakhir. Ilustrasi sequential search ini dapat dilihat pada Gambar 1

![gambar](sequnt.png)


Sequential search (pencarian berurutan) adalah cara mencari data dengan memeriksa satu per satu dari awal sampai ketemu (atau sampai habis).

Cara kerjanya:

1. Mulai dari elemen pertama
   
2. Bandingkan dengan nilai yang dicari

 
3. Kalau cocok → berhenti, kembalikan posisinya

   
4. Kalau tidak cocok → lanjut ke elemen berikutnya

   
5. Kalau sudah habis dan tidak ketemu → kembalikan -1

Contoh sederhana:

Data: [5, 3, 8, 1, 9], cari angka 8

Cek index 0 → 5  ≠ 8, lanjut

Cek index 1 → 3  ≠ 8, lanjut

Cek index 2 → 8  = 8, KETEMU! → return index 2


Karakteristik sequential search:

Keterangan:

Terbaik → langsung ketemu di index 0 → 1 kali cek

Teburuk → Ada di akhir atau tidak ada → n kali cek

Rata-rata → n/2 kali cek

Cocok untul → Data kecil, data tidak terurut

Tidak cocok untuk → Data besar(lambat)

sequential search itu simpel tapi tidak efisien untuk data besar. Kalau datanya sudah terurut, lebih baik pakai binary search yang jauh lebih cepat.


contoh kode sederhana :

```python
def sequential_search(data, target):
    for i in range(len(data)):
        print(f"Cek index {i} → {data[i]}", end="")
        if data[i] == target:
            print(f"  ✓ KETEMU!")
            return i
        else:
            print(f"  ✗ lanjut...")
    print("Data tidak ditemukan")
    return -1

data = [5, 3, 8, 1, 9]
target = 8

print(f"Data: {data}")
print(f"Mencari: {target}\n")
hasil = sequential_search(data, target)
print(f"\nHasil: index ke-{hasil}")
```

Unordered List Sequential Search

Pada unordered list sequential search, data berada pada keadaan acak, tidak terurut, sehingga
pencarian harus dilakukan mulai dari indeks awal sampai indeks terakhir dari data, atau pencarian
berhenti ketika data sudah ditemukan

Unordered List Sequential Search adalah sequential search yang diterapkan pada data yang tidak terurut (acak).

Karakteristiknya:

Karena datanya acak, tidak ada cara lain selain memeriksa satu per satu dari awal sampai akhir — tidak bisa skip atau potong jalan.

Cara kerjanya:

1. Mulai dari index pertama

2. Bandingkan tiap elemen dengan target

3. Kalau cocok → kembalikan indexnya

4. Kalau sudah sampai akhir dan tidak ketemu → kembalikan -1

Analisis kompleksitas:

Kondisi → Terbaik  Jumlah perbandingan → 1 kali(target di index pertama)

Kondisi → Terburuk Jumlah perbandingan → n kali(target di akhir atau tidak ada)

Kondisi → Rata-rata Jumlah perbandingan → n/2 kali


contoh kode sederhana :

```python
def unordered_sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

data = [20, 5, 13, 1, 8, 17, 3] 
target = 8

hasil = unordered_sequential_search(data, target)

if hasil != -1:
    print(f"Data {target} ditemukan di index {hasil}")
else:
    print(f"Data {target} tidak ditemukan")
```

unordered list sequential search tidak bisa dioptimasi lebih jauh karena urutan datanya acak — satu-satunya cara ya memeriksa semua elemen satu per satu.



code di bawah ini implementasi algoritma unordored list sequential search.

```python
def seqSearch(listData, data):
    ind = 0
    found = False
    while ind < len(listData) and not found:
        if listData[ind] == data:
            found = True
        else:
            ind = ind+1
    return found
```

Ordered List Sequential Search


Ordered List Sequential Search adalah sequential search yang diterapkan pada data yang sudah terurut (dari kecil ke besar).

Keunggulan dibanding Unordered:

Karena datanya sudah terurut, pencarian bisa berhenti lebih awal tanpa harus memeriksa semua elemen. Kalau elemen yang sedang diperiksa sudah lebih besar dari target, berarti target pasti tidak ada — tidak perlu lanjut.


Cara kerjanya:

1. Mulai dari index pertama

2. Bandingkan tiap elemen dengan target

3. Kalau cocok → kembalikan indexnya

4. Kalau elemen saat ini lebih besar dari target → berhenti, kembalikan -1

5. Kalau sudah sampai akhir dan tidak ketemu → kembalikan -1

Contoh:

Data: [1, 3, 5, 8, 13, 17, 20], cari angka 6

Cek index 0 → 1  < 6, lanjut

Cek index 1 → 3  < 6, lanjut

Cek index 2 → 5  < 6, lanjut

Cek index 3 → 8  > 6, BERHENTI → data tidak ada

Tanpa perlu cek index 4, 5, 6 — lebih efisien!


dibawah ini kode sederhana :

```python
def ordered_sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
        elif data[i] > target:
            return 'Data tidak ada'
    return 'Data tidak ada'

data = [1, 3, 5, 8, 13, 17, 20]
target = 6

hasil = ordered_sequential_search(data, target)
print(f"Hasil pencarian: {hasil}")
```

Secara teori kompleksitasnya sama, tapi dalam praktiknya ordered lebih cepat karena bisa berhenti di tengah jalan saat target sudah pasti tidak ada.

Pada ordered list sequential search, data sudah dalam keadaan terurut, hal ini tentunya dapat

Berikut implementasi algoritma sequential search pada ordered list

```python
def orderedSeqSearch(listData, data):
    ind = 0
    found = False
    stop = False
    position=[]
    while ind < len (listData) and not found and not stop:
        if listData[ind] == data:
            position.append(ind)
        else:
            if listData[ind] > data:
                stop = True
            else:
                ind = ind+1

    if found:
        return ind
    else:
        return ('Data tidak ada')       
```

Binary Serach

Binary Search adalah algoritma pencarian yang bekerja dengan cara membagi dua ruang pencarian setiap langkahnya.
Syarat utamanya: data harus sudah terurut.

Cara kerjanya:

1. Tentukan titik tengah array

2. Bandingkan nilai tengah dengan target

3. Jika sama → ketemu!

4. Jika target lebih kecil → cari di bagian kiri

5. Jika target lebih besar → cari di bagian kanan

6. Ulangi sampai ketemu atau habis

Contoh sederhana:

Array: [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
Target: 23

Langkah 1: tengah = 16 → 23 > 16, cari kanan

Langkah 2: tengah = 38 → 23 < 38, cari kiri

Langkah 3: tengah = 23 → ketemu!

Hanya 3 langkah dari 10 elemen.

Kenapa ini efisien?

Setiap langkah membuang setengah dari sisa data. Untuk 1.000 elemen, paling banyak butuh ~10 langkah. Untuk 1.000.000 elemen, hanya ~20 langkah. Ini yang disebut kompleksitas O(log n).
Bandingkan dengan linear search yang harus cek satu per satu — bisa butuh 1.000.000 langkah untuk 1.000.000 elemen.

Kapan tidak bisa dipakai:

Jika datanya belum terurut, Binary Search tidak bisa digunakan. Harus sort dulu, atau gunakan linear search.

Pencarian dengan algoritma sequential search akan menemukan data dengan cepat jika data
yang dicari terletak di indeks pertama dari suatu list data. Akan tetapi jika data yang dicari berada
di posisi terakhir dari suatu list atau posisi lain, maka algoritma sequential search ini akan
membutuhkan waktu yang lebih lama, karena algoritma ini akan mencari satu persatu, dimulai dari
indeks pertama sampai dengan indeks terakhir.

Pada algoritma binary search pencarian tidak dilakukan satu persatu, melainkan memanfaatkan
kelebihan pencarian di data yang sudah terurut.
Pada binary search, pencarian dilakukan pertama kali di indeks list yang berada ditengah.
Jika data yang dicari sama dengan data pada indeks tersebut, maka pencarian berakhir. Akan
tetapi jika data yang dicari lebih besar dari data yang berada di indeks tengah tersebuh, maka
dilakukan pencarian lagi. Hanya saja pencarian tidak dilakukan pada seluruh data, hanya data
yang berada indeks-indeks setelah indeks tengah tadi, karena data sebelum indeks tengah, pasti
lebih kecil nilainya, sehingga tidak perlu dilakukan pencarian lagi. Proses ini dilakukan secara
terus menerus sampai data ketemu atau tidak ada lagi data yang dapat dicari di dalam list. Oleh
karena itu, binary search tidak perlu mencari di semua data pada list, sehingga waktu komputasi
yang dibutuhkan lebih cepat dibandingkan dengan pencarian sequential search.

Ilustrasi pencarian binary search ini dapat dilihat pada Gambar berikut.

![gambar](start.png)


Dibawah ini code untuk binary search dengan metode iteratif

```python
def binarySearch(listData, data):
    first = 0
    last = len(listData) - 1
    found = False

    while first <= last and not found:
        midpoint = (first + last) // 2
        if listData[midpoint] == data:
            found = True

        else:
            if data < listData[midpoint]:
                last = midpoint - 1
            else:
                first = midpoint + 1

    return found
```

```python
a=[4,6,10,34,56,78,99]
print(binarySearch(a, 34))
```

```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

arr = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]

target = 23
result = binary_search(arr, target)

if result != -1:
    print(f"Data {target} ditemukan pada indeks: {result}")
    print(f"Verifikasi: arr[{result}] = {arr[result]}")
else:
    print(f"Data {target} tidak ditemukan")
```

Hashing adalah teknik untuk mengubah data (angka, string, dll) menjadi sebuah indeks menggunakan fungsi tertentu, sehingga data bisa disimpan dan dicari dengan sangat cepat.

Konsep dasarnya:

Daripada mencari data satu per satu atau pakai binary search, hashing langsung "menghitung" di mana posisi data tersebut disimpan. Seperti punya "rumus" untuk langsung tahu lokasi barang tanpa harus mencari.

Fungsi Hash:

Fungsi hash menerima data dan menghasilkan indeks. Contoh paling sederhana: indeks = data % ukuran_array. Misalnya array berukuran 10, data 23 → 23 % 10 = 3, jadi data 23 disimpan di indeks 3. Saat dicari, langsung hitung lagi 23 % 10 = 3, langsung menuju indeks 3.

Collision:
Masalah muncul ketika dua data menghasilkan indeks yang sama. Misalnya data 23 dan 33 sama-sama menghasilkan indeks 3. Ini disebut collision. Ada beberapa cara mengatasinya, yang paling umum adalah chaining (setiap slot menyimpan list/linked list) dan open addressing (cari slot kosong berikutnya).

Kenapa hashing cepat:

Pencarian, penyisipan, dan penghapusan data rata-rata hanya butuh O(1) — waktu konstan, tidak peduli seberapa besar datanya. Ini lebih cepat dari binary search yang O(log n).

Penggunaan nyata:

Marge Sort

Merge Sort adalah algoritma pengurutan yang menggunakan pendekatan divide and conquer — pecah, urutkan, lalu gabungkan.

Cara kerjanya ada 3 tahap:

1. Divide (Bagi)
Array dipecah terus menjadi dua bagian yang sama besar, sampai setiap bagian hanya tersisa 1 elemen. Satu elemen sudah pasti "terurut" dengan sendirinya.

2. Conquer (Urutkan)
Bagian-bagian kecil tadi mulai digabungkan dua-dua. Saat menggabungkan, elemen dibandingkan satu per satu dan disusun dari yang terkecil.


3. Merge (Gabungkan)
Terus digabungkan ke atas sampai menjadi satu array utuh yang sudah terurut.

Contoh visual dengan array [38, 27, 43, 3]:

![gambar](visual.png)

Karakteristik Merge Sort:

- Time complexity: O(n log n) — di semua kasus (best, average, worst). Inilah keunggulan utamanya dibanding Bubble Sort atau Insertion Sort yang bisa O(n²)

- Space complexity: O(n) — butuh array tambahan saat proses merge

- Stable sort — urutan elemen yang nilainya sama tidak berubah

- Tidak bergantung data awal — mau datanya sudah hampir terurut atau acak total, tetap O(n log n)

Merge Sort cocok untuk data berukuran besar, atau ketika stabilitas urutan penting. Kelemahannya hanya satu: butuh memori ekstra karena harus membuat array sementara saat menggabungkan.


Permasalahan sorting akan lebih mudah dilaksanakan jika hanya terdapat satu elemen saja yang

Jika hanya terdapat satu elemen dalam sebuah list data, maka data tersebut otomatis sudah
dalam keadaan terurut.

Pada Gambar tersebut ditunjukka, pada proses merging, data yang terdapat pada list yang akan

Code
Berikut adalah code untuk algoritma pengurutan Merge Sort, dan contoh pemanggilan fungsi
merge sort ini.

```python
def merge_sort(a_list):
    print("Splitting ", a_list)
    if len(a_list) > 1:
        mid = len(a_list) // 2
        left_half = a_list[:mid]
        right_half = a_list[mid:]
        merge_sort(left_half)
        merge_sort(right_half)
        i = 0 #indeks left half
        j = 0 #indeks right half
        k = 0 # indeks hasil merging/sorting
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                a_list[k] = left_half[i]
                i = i + 1
            else:
                a_list[k] = right_half[j]
                j = j + 1
            k = k + 1
        while i < len(left_half):
            a_list[k] = left_half[i]
            i = i + 1
            k = k + 1
        while j < len(right_half):
            a_list[k] = right_half[j]
            j = j + 1
            k = k + 1

    print("Merging ", a_list)
```

Merge Sort + Hitung Inversi

Buatlah fungsi Merge Sort yang tidak hanya mengurutkan array, tetapi juga menghitung jumlah inversi.

Inversi adalah pasangan (i, j) dimana i < j tetapi arr[i] > arr[j]
Contoh: pada array [3, 1, 2], inversinya adalah (3,1) dan (3,2) → total 2 inversi

Gunakan array ini:
[6, 5, 3, 1, 8, 7, 2, 4]
Tentukan:

1. Array setelah diurutkan
2. Total jumlah inversi
3. Sebutkan semua pasangan inversinya




```python
def merge_count(left, right):
    result = []
    inversions = 0
    i = 0
    j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            inversions += len(left) - i
            result.append(right[j])
            j += 1

    result += left[i:]
    result += right[j:]
    return result, inversions


def merge_sort_count(arr):
    if len(arr) <= 1:
        return arr, 0

    mid = len(arr) // 2

    left,  inv_left  = merge_sort_count(arr[:mid])
    right, inv_right = merge_sort_count(arr[mid:])

    sorted_arr, inv_split = merge_count(left, right)

    total_inv = inv_left + inv_right + inv_split
    return sorted_arr, total_inv


arr = [6, 5, 3, 1, 8, 7, 2, 4]
print("Array asal :", arr)

hasil, total = merge_sort_count(arr)
print("Terurut    :", hasil)
print("Total inversi:", total)

print("\nSemua pasangan inversi:")
pasangan = []
for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] > arr[j]:
            pasangan.append((arr[i], arr[j]))
            print(f"  ({arr[i]}, {arr[j]})")

print(f"\nTotal pasangan: {len(pasangan)}")
```

- Semua elemen lebih besar dari pivot → pindah ke kanan
- Pivot berada di posisi yang tepat (tidak akan bergerak lagi)

Algoritma untuk membagi data atau partisi data menjadi dua bagian berdasarkan nilai pivot

Pada algoritma tersebut, nilai pivot merupakan data yang terdapat pada indeks awal dari data

Code

Berikut fungsi untuk partisi data, dengan tiga parameter, yaitu A adalah data , p adalah indeks
awal dari data yang akan dipartisi, dan q adalah indeks terakhir dari data yang akan dipartisi

```python
def partition (A,p,q):
    x=A[p]
    i=p
    for j in range(p+1, q+1) :
        if A[j]<=x:
            i=i+1
            A[i],A[j]=A[j],A[i]
        print(A)
    A[p],A[i]=A[i],A[p]
    return i
```

Tahap kedua pada algoritma quicsort adalah tahapan combine. Pada tahapan ini, hanya

Code

Berikut adalah fungsi combine untuk algoritma quicksort

```python
def Quicksort(A,p,q):
    if p<q:
        r=partition(A,p,q)
        Quicksort(A, p, r-1)
        Quicksort(A, r+1, q)
```

Latihan Quick Sort + Lacak Posisi Pivot

Buatlah fungsi Quick Sort yang tidak hanya mengurutkan array, tetapi juga mencatat di langkah ke berapa setiap pivot menemukan posisi finalnya.

Gunakan array ini:
[10, 80, 30, 90, 40, 50, 70]

Tentukan:

1. Array setelah diurutkan
   
2. Urutan pivot yang terpilih

3. Di langkah ke berapa tiap pivot "terkunci" di pos

```python
langkah = 0 

def partisi(arr, low, high):
    global langkah
    pivot = arr[high]
    i = low - 1  

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i] 

   
    arr[i + 1], arr[high] = arr[high], arr[i + 1]

    langkah += 1
    posisi_final = i + 1
    print(f"  Langkah {langkah}: pivot={pivot} terkunci di indeks {posisi_final} → {arr}")

    return posisi_final


def quick_sort(arr, low, high):
    if low < high:
        pos = partisi(arr, low, high)

        quick_sort(arr, low, pos - 1)
        quick_sort(arr, pos + 1, high)



arr = [10, 80, 30, 90, 40, 50, 70]
print("Sebelum :", arr)
print()
print("=== Proses Partisi ===")

quick_sort(arr, 0, len(arr) - 1)

print()
print("Sesudah :", arr)
```

Counting Sort

Counting Sort adalah algoritma pengurutan yang bekerja dengan cara menghitung kemunculan setiap elemen, bukan dengan membandingkan elemen satu sama lain seperti Quick Sort atau Merge Sort.

Cara kerjanya ada 3 tahap:

1. Hitung (Count)
Buat array count berukuran sebesar nilai maksimum+1. Lalu hitung berapa kali setiap angka muncul di array asal — indeks array count = nilai elemennya.
2. Kumulatif (Cumulative)
Ubah array count menjadi kumulatif — setiap posisi ditambah dengan nilai sebelumnya. Ini berguna untuk menentukan posisi final setiap elemen di array hasil.
3. Bangun hasil (Output)
Iterasi array asal dari kanan ke kiri, tempatkan setiap elemen ke posisi yang sudah ditentukan oleh array kumulatif, lalu kurangi count-nya.

Contoh visual dengan array [4, 2, 2, 8, 3, 3, 1]:

![gambar](counting.png)

Counting Sort sangat cepat ketika nilai k (nilai maksimum) tidak terlalu jauh dari n (jumlah elemen). Misalnya mengurutkan nilai ujian 100 siswa (nilai 0–100) — k=100, n=100, sangat efisien. Sebaliknya, kalau array berisi angka sampai jutaan tapi elemennya sedikit, array count-nya jadi sangat besar dan boros memori.

Algoritma sorting yang ketiga adalah algoritma sorting yang memiliki waktu komputasi lebih rendah
dibandingkan kedua algoritma sebelumnya (merge sort dan quick sort), yaitu counting sort.

Pada algoritma ini, masing-masing data yang akan diurutkan, dihitung jumlahnya terlebih dahulu.
Sehingga ketika jumlah masing-masing data telah diketahui, maka data-data ini akan mudah
ditempatkan ke dalam indeks masing-masing.

Misalkan terdapat data [4, 3, 1, 2, 4, 2, 1, 1], maka akan dilakukan terlebih dahulu, perhitungan
jumlah masing-masing data, yaitu :
data-'1' berjumlah tiga data, sehingga akan menempati posisi indeks ke-0 sampai dengan indeks
ke-2
data-'2' berjumlah dua buah data, sehingga akan menempati indeks ke-3 sampai dengan ke-4.
data-'3' berjumlah satu buah data, sehingga hanya akan menempati indeks ke-5, sedangkan
data-'4' berjumlah dua buah data, dan kedua data ini akan menempati indeks ke-6 dan ke-7.

Code

Berikut adalah code untuk algoritma counting sort.
Loop pertama untuk menyiapkan variabel yang berfungsi menyimpan jumlah komulatif dari
masing-masing data, sehingga dengan nilai komulatif ini, maka masing-masing data dapat
diletakkan pada indeks-indeks yang tepat berdasarkan nilai komulatif ini.
Loop kedua digunakan untuk menghitung jumlah elemen pada masing-masing data. Misalkan
pada contoh sebelumnya, data '1' berjumlah tiga buah elemen.
Dari loop kedua ini, dilakukan perhitungan komulatif dari masing-masing data.
Loop terakhir akan data akan ditempatkan sesuai dengan nilai komulatif tersebut.

```python
def countingSort(A,k):
    C=[]
    B=[]
    for i in range(k+1):
        C.append(0)

    for j in range(len(A)):
        C[A[j]]=C[A[j]]+1
        B.append(0)

    for i in range(1,k+1):
        C[i]=C[i]+C[i-1]

    for j in range(len(A)-1,-1,-1):
        B[C[A[j]]-1]=A[j]
        C[A[j]]=C[A[j]]-1
    return(B)  
```

```python
a_list = [54, 26, 93, 17, 1, 31, 2, 55, 20,93,20,3,4,5]
countingSort(a_list,93)
```

Counting Sort 2D: Urutkan Jadwal Berdasarkan JamKamu diberikan data jadwal kereta dalam bentuk (kode_kereta, jam_berangkat, tujuan). Gunakan Counting Sort untuk mengurutkan berdasarkan jam_berangkat, lalu:

1. Tampilkan jadwal terurut per jam
2. Hitung kepadatan per sesi (pagi 05–11, siang 12–16, malam 17–23)
3. Temukan jam tersibuk (jam dengan kereta terbanyak)
4. Jika jam sama, urutkan berdasarkan kode kereta secara alfabetis


Gunakan data ini:

jadwal = [
    ("KA-101", 8,  "Surabaya"),
    
    ("KA-205", 14, "Bandung"),
    
    ("KA-303", 8,  "Yogyakarta"),  # jam sama KA-101
    
    ("KA-412", 6,  "Malang"),
    
    ("KA-118", 20, "Jakarta"),
    
    ("KA-220", 14, "Semarang"),    # jam sama KA-205
    
    ("KA-307", 6,  "Bali"),        # jam sama KA-412
    
    ("KA-500", 23, "Medan"),
    
    ("KA-102", 8,  "Surabaya"),    # jam sama KA-101 & KA-303
    
    ("KA-410", 17, "Makassar"),
]


```python
def counting_sort_jadwal(data):
    # Ambil range jam (0-23)
    min_jam = min(jam for _, jam, _ in data)
    max_jam = max(jam for _, jam, _ in data)
    k = max_jam - min_jam + 1

    # Tahap 1: Hitung kemunculan tiap jam
    count = [0] * k
    for _, jam, _ in data:
        count[jam - min_jam] += 1

    print("=== Tahap 1: Hitung per Jam ===")
    for i, c in enumerate(count):
        if c > 0:
            print(f"  jam {i + min_jam:02d}:00 → {c} kereta")

    # Tahap 2: Kumulatif
    for i in range(1, k):
        count[i] += count[i - 1]

    print("\n=== Tahap 2: Kumulatif ===")
    for i, c in enumerate(count):
        if c > 0:
            print(f"  count[{i + min_jam:02d}] = {c}")

    # Tahap 3: Bangun output (reversed untuk stable)
    output = [None] * len(data)
    for item in reversed(data):
        _, jam, _ = item
        posisi = count[jam - min_jam] - 1
        output[posisi] = item
        count[jam - min_jam] -= 1

    return output


def urutkan_jam_sama(data):
    """Jika jam sama, urutkan kode kereta alfabetis (insertion sort)"""
    arr = list(data)
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j][1] == key[1] and arr[j][0] > key[0]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr


def sesi(jam):
    if 5 <= jam <= 11:
        return "Pagi"
    elif 12 <= jam <= 16:
        return "Siang"
    else:
        return "Malam"


jadwal = [
    ("KA-101", 8,  "Surabaya"),
    ("KA-205", 14, "Bandung"),
    ("KA-303", 8,  "Yogyakarta"),
    ("KA-412", 6,  "Malang"),
    ("KA-118", 20, "Jakarta"),
    ("KA-220", 14, "Semarang"),
    ("KA-307", 6,  "Bali"),
    ("KA-500", 23, "Medan"),
    ("KA-102", 8,  "Surabaya"),
    ("KA-410", 17, "Makassar"),
]

print("=== Data Jadwal Awal ===")
for kode, jam, tujuan in jadwal:
    print(f"  {kode}  jam {jam:02d}:00  → {tujuan}")

print()
hasil = counting_sort_jadwal(jadwal)
hasil = urutkan_jam_sama(hasil)

# Tampilkan jadwal terurut
print("\n=== Jadwal Terurut ===")
print(f"{'No':<4} {'Kode':<10} {'Jam':<8} {'Tujuan':<12} {'Sesi'}")
print("-" * 44)
for i, (kode, jam, tujuan) in enumerate(hasil, 1):
    print(f"  {i:<3} {kode:<10} {jam:02d}:00   {tujuan:<12} {sesi(jam)}")

# Kepadatan per sesi
print("\n=== Kepadatan per Sesi ===")
sesi_data = {"Pagi": [], "Siang": [], "Malam": []}
for kode, jam, tujuan in hasil:
    sesi_data[sesi(jam)].append(kode)

for s, kereta in sesi_data.items():
    bar = "█" * len(kereta)
    print(f"  {s:<6} ({len(kereta)} kereta) {bar}")
    for k in kereta:
        print(f"           └ {k}")

# Jam tersibuk
print("\n=== Jam Tersibuk ===")
jam_count = {}
for _, jam, _ in hasil:
    jam_count[jam] = jam_count.get(jam, 0) + 1

max_count = max(jam_count.values())
for jam, count in sorted(jam_count.items()):
    flag = " ← TERSIBUK" if count == max_count else ""
    bar  = "█" * count
    print(f"  jam {jam:02d}:00 → {bar} ({count} kereta){flag}")
```

Override Method
Override Method merupakan penambahan fungsi-fungsi pada syntax-syntax yang sudah ada.
Seperti yang dijelaskan sebelumnya, untuk menampilkan property yang terdapat pada suatu
obyek, tidak dapat dilakukan dengan menggunakan perintah print() .

Cara pertama yang sudah dijelaskan adalah membuat method baru di dalam class, yang berfungsi
untuk menampilkan property suatu class.
Cara kedua adalah dengan cara override method fungsi yang sudah tersedia oleh bahasa
pemrograman. Syntax yang akan dioverride adalah syntax print() .
Yang perlu dilakukan agar suatu object dapat dipanggil dengan perintah print adalah object
tersebut harus dirubah menjadi string. Untuk merubah suatu variabel menjadi string, maka suatu
class dilengkapi dengan method __str__ . Hanya saja method ini tidak bisa berfungsi secara
sempurna sesuai dengan apa yang diinginkan oleh programmer, yaitu agar syntax print dapat
digunakan langsung untuk menampilkan property yang terdapat pada suatu class.

Code

Berikut contoh override method __str__ pada class BilanganKompleks

```python
class BilanganKompleks:
    def __init__(self,a,b):
        self.real=a
        self.im=b
    def display(self):
        print (self.real,"+",self.im,"i")
    def __str__(self):
        return str(self.real)+" + "+str(self.im)+" i "
```

```python
data1=BilanganKompleks(4,6)
data2=BilanganKompleks(2,5)
```

```python
print('Override Method')
print(data1)
print(data2)
print('Method dalam Class')
data1.display()
data2.display()
```

Penjumlahan dua buah bilangan kompleks berbeda dengan penjumlahan bilangan yang lain,
karena bilangan kompleks ini terdiri dari dua bagian yaitu real dan imajiner. Untuk menjumlahkan
bilangan kompleks, maka bagian real harus dijumlahkan pada bagian real juga dari bilangan lain,
dan bagian imajiner harus ditambahkan dengan bagian imajiner, seperti persamaan berikut:

                     x1 = x1 + y1 i
                     x2 = x2 + y2 i
                x1 + x2 = (x1 + y1 i) + (x2 + y2i)
                        =(x1 + x2)+((y1 + y2)i)

Seperti halnya fungsi untuk menampilkan property yang terdapat pada suatu class, maka untuk
melakukan operasi penjumlahan ini juga dapat dilakukan dengan dua buah cara, yaitu :

- Membuat metode di dalam class untuk operasi penjumlahan
- Override method

Code

Contoh code berikut adalah pembuatan method baru di dalam class, yang berfungsi untuk
menjumlahkan dua buah bilangan kompleks.

```python
class BilanganKompleks:
    def __init__(self,a,b):
        self.real=a
        self.im=b
    def display(self):
        print (self.real,'+',self.im,'i')
    def __str__(self):
        return str(self.real) + " + " + str(self.im) + " i "
    def addKompleks(self,obj):
        a=self.real+obj.real
        b=self.im+obj.im
        return BilanganKompleks(a,b)
```

```python
data1=BilanganKompleks(4,6)
data2=BilanganKompleks(2,5)
```

```python
jumlah=data1.addKompleks(data2)
data1.display()
data2.display()
print(jumlah)
```
